```python
class Value:
    def __init__(self, data, _children=None, _op=None):
        self.data = data
        self._children = _children
        self._prev = set(_children) if _children else set()
        self._op = _op

    def __repr__(self):
        return f"Value(data={self.data})"

    def __add__(self, other):
        data = self.data + other.data
        out = Value(
            data=data,
            _children=(self, other),
            _op="+"
        )
        return out

    def __mul__(self, other):
        data = self.data * other.data
        out = Value(
            data=data,
            _children=(self, other),
            _op="*"
        )
        return out

    # @staticmethod
    # def create_op(data, _op):
    #     out = Value(
    #         data=data,
            
    #     )
    #     return Out

    # def __radd__(self, other):
    #     return self + other
```
